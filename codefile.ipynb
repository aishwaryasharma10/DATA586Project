```python
import pandas as pd
import numpy as np
```

```python
def logparser(path = 'HDFS.log'):
    with open(path) as f:
        data_list = list()
        for line in f.readlines():
            line = line.split(":", 1)
            observation = line[0].split() + [line[1].strip()]
            data_list.append(observation)
        return pd.DataFrame(data_list, columns = ["date", "timestamp", "id", "state", "component", "message"])
```

```python
df = logparser()
```

```python
df.to_csv('data', index=False)
```
